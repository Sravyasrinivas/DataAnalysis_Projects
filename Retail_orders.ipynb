```python
import pandas as pd
df = pd.read_csv("orders.csv",na_values=['Not Available','unknown'])
df['Ship Mode'].unique()
```

```python
#rename column names and make them lower case and replace space with underscore
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
#df.columns
df.head(5)
```

```python
#derive new column discount, sale price and profit

df['discount'] = df['list_price'] * df['discount_percent']/100
#df['discount']
df['sale_price'] = df['list_price'] - df['discount']
#df['sale_price']
df['profit'] = df['sale_price'] - df['cost_price']
df.head(5)
```

```python
#convert order date from object data type to datetime

df['order_date'] = pd.to_datetime(df['order_date'],format='%Y-%m-%d')
#df.info() or df.dtype()
```

The error you're encountering typically occurs when you assign the result of an operation that modifies a DataFrame in place (inplace=True) to a variable, leading the variable to become None. The drop method with inplace=True modifies the DataFrame in place and returns None, so if you assign it back to df, df will become None.

```python
#drop cost_price,list_price and discount_percent columns

df.drop(columns=['cost_price','list_price','discount_percent'],inplace = True)
df.head(5)
```
